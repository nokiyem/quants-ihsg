```python
from selenium.webdriver.firefox.options import Options as FirefoxOptions
import pandas as pd

na_values1=['-1.#IND', '1.#QNAN', '1.#IND', '-1.#QNAN', '#N/A N/A', '#N/A','NA', 
           '', '#NA', 'NULL', 'null', 'NaN', '-NaN', 'nan', '-nan', '','-']

df1 = pd.read_csv('all_saham_scrape.csv',thousands=',',na_values=na_values1)
df1.head()
```

```python
col_list=df1.columns.tolist()
for i in col_list:
    if '12M' in i:
        #print(str(i[4:]))
        newcol = str(i[4:])
        df1.rename(columns = {i:newcol}, inplace = True) 
```

```python
from selenium.webdriver.firefox.options import Options as FirefoxOptions
import pandas as pd

df_id = pd.read_csv('saham-scrape.csv')
df_id.head()
df_all = pd.DataFrame() 

for index, row in df_id.iterrows(): #loop sheets
    df_aali = df1.loc[df1['Saham']==row["NAME"]]
    df_aali = df_aali.drop('Saham', 1)
    df_aali = df_aali.sort_values(by=['Indikator'])


    df_T = df_aali.T
    #replace columns
    df_T.columns = df_T.iloc[0]
    #remove duplicate rows
    df_T = df_T.iloc[1:]
    
    #print(df_T.columns)
    
    #df_T.head(5)
    df_T["Saham"] = row["NAME"]
    df_T['Tahun'] = df_T.index
    df_T['Tahun'] = df_T['Tahun'].astype(int) 
    df_T['Tahun'] = df_T['Tahun']+1
    df_T.index = df_T["Saham"]
    
    # Remove column name 'Saham' 
    df_T = df_T.drop(['Saham'], axis = 1) 
    #remove duplicate columns
    df_T = df_T.loc[:,~df_T.columns.duplicated()]


    ##df_T.info()
    df_all = pd.concat([df_all,df_T])
    
```

```python
df_harga =pd.read_excel("data/learning-capital-cleansing.xlsx")
df_harga.head()
```

```python
df_harga['Tahun']=df_harga['Tahun'].astype(int) 
#df_harga = df_harga.convert_dtypes()
#df_all = df_all.convert_dtypes()

df_join = pd.merge(df_all, df_harga, how ='inner', on =['Saham','Tahun']) 
df_join.head()
```

```python
col_list=df_join.columns.tolist()

for i in col_list:
    if i not in ["Saham","Tahun","CLOSING OCT"]:
        df_join[i] = df_join[i].astype(str)
        #print(df_all[i])
        df_join[i]=df_join[i].str.replace(',', '').replace('%', '')

for i in col_list:
    if i not in ["Saham","Tahun","CLOSING OCT"]:
        df_join[i] = df_join[i].astype(str)
        #print(df_all[i])
        df_join[i]=df_join[i].str.replace('%', '').astype(float)
```

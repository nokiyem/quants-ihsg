```python
col_list=df1.columns.tolist()
for i in col_list:
    if 'Q' in i:
        #print(str(i[3:])+str(i[:2][1:]))
        newcol = str(i[3:])+str(i[:2][1:])
        df1.rename(columns = {i:newcol}, inplace = True) 
```

```python
from selenium.webdriver.firefox.options import Options as FirefoxOptions
import pandas as pd

df_id = pd.read_csv('saham-scrape.csv')
df_id.head()
df_all = pd.DataFrame() 

for index, row in df_id.iterrows(): #loop sheets
    df_aali = df1.loc[df1['Saham']==row["NAME"]]
    df_aali = df_aali.drop('Saham', 1)
    df_aali = df_aali.sort_values(by=['Dalam IDR'])


    df_T = df_aali.T
    #replace columns
    df_T.columns = df_T.iloc[0]
    #remove duplicate rows
    df_T = df_T.iloc[1:]
    
    #print(df_T.columns)
    
    #df_T.head(5)
    df_T["Saham"] = row["NAME"]
    df_T['Quarter'] = df_T.index
    df_T.index = df_T["Saham"]
    #remove duplicate columns
    df_T = df_T.loc[:,~df_T.columns.duplicated()]


    ##df_T.info()
    df_all = pd.concat([df_all,df_T])
    
```

```python
df_all
```

```python
df_all = df_all.apply(lambda x: x.fillna(0))
```

```python
df_all['Capital expenditure (Quarter)'] = df_all['Capital expenditure (Quarter)'].astype(str).str.replace('\((.*)\)', '-\\1')
```

```python
nama_file="all_saham_scrape_quarterly1_transpose.csv"
df_all.to_csv(nama_file) 
```

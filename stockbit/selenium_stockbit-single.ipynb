```python
import time
import pytest
import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
```

```python

from selenium.webdriver.firefox.options import Options as FirefoxOptions
import pandas as pd

driver = webdriver.Chrome("D:\Downloads\chromedriver.exe")

vars = {}

driver.get("https://stockbit.com/")
time.sleep(2)
driver.find_element(By.CSS_SELECTOR, ".desktop .login-ldn").click()
time.sleep(4)
driver.find_element(By.ID, "username").send_keys("username")
driver.find_element(By.ID, "password").send_keys("password")
driver.find_element(By.ID, "loginbutton").click()
time.sleep(2)

driver.find_element(By.CSS_SELECTOR, ".searchbox").click()
driver.find_element(By.CSS_SELECTOR, ".searchbox").send_keys("AALI")
driver.find_element(By.CSS_SELECTOR, ".searchbox").send_keys(Keys.ENTER)
driver.execute_script("window.scrollTo(0,417)")
time.sleep(4)
driver.find_element(By.CSS_SELECTOR, ".navnav:nth-child(5) > b").click()
time.sleep(4)
dropdown = driver.find_element(By.NAME, "currency")
time.sleep(1)
dropdown.find_element(By.XPATH, "//option[. = 'IDR']").click()
time.sleep(1)
dropdown = driver.find_element(By.NAME, "ordertype")
time.sleep(1)
dropdown.find_element(By.XPATH, "//option[. = 'ID']").click()
time.sleep(2)

dropdown = driver.find_element(By.NAME, "statementtype")
dropdown.find_element(By.XPATH, "//option[. = 'Annual']").click()
time.sleep(2)

ratio1 = driver.find_element_by_xpath('//div[@id="keyratio"]')
table1=pd.read_html(str(ratio1.get_attribute('innerHTML')))

driver.find_element(By.NAME, "reporttype").click()
dropdown = driver.find_element(By.NAME, "reporttype")
dropdown.find_element(By.XPATH, "//option[. = 'Cash Flow']").click()
time.sleep(2)
driver.find_element(By.NAME, "reporttype").click()
dropdown = driver.find_element(By.NAME, "currency")
dropdown.find_element(By.XPATH, "//option[. = 'IDR']").click()
dropdown = driver.find_element(By.NAME, "ordertype")
dropdown.find_element(By.XPATH, "//option[. = 'ID']").click()
#driver.find_element(By.LINK_TEXT, "Key Ratio").click()
ratio2 = driver.find_element_by_xpath('//div[@id="keyratio"]')
table2=pd.read_html(str(ratio2.get_attribute('innerHTML')))

driver.find_element(By.NAME, "reporttype").click()
dropdown = driver.find_element(By.NAME, "reporttype")
dropdown.find_element(By.XPATH, "//option[. = 'Balance Sheet']").click()
time.sleep(2)

driver.find_element(By.NAME, "reporttype").click()
dropdown = driver.find_element(By.NAME, "currency")
dropdown.find_element(By.XPATH, "//option[. = 'IDR']").click()
dropdown = driver.find_element(By.NAME, "ordertype")
dropdown.find_element(By.XPATH, "//option[. = 'ID']").click()
driver.find_element(By.LINK_TEXT, "Key Ratio").click()
time.sleep(2)
ratio3 = driver.find_element_by_xpath('//div[@id="keyratio"]')
table3=pd.read_html(str(ratio3.get_attribute('innerHTML')))


driver.find_element(By.LINK_TEXT, "Financial Health").click()
time.sleep(2)
ratio4 = driver.find_element_by_xpath('//div[@id="financialhealth"]')
table4=pd.read_html(str(ratio4.get_attribute('innerHTML')))

driver.find_element(By.LINK_TEXT, "Efficiency").click()
time.sleep(2)
ratio5 = driver.find_element_by_xpath('//div[@id="efficiency"]')
table5=pd.read_html(str(ratio5.get_attribute('innerHTML')))

driver.find_element(By.CSS_SELECTOR, ".ant-dropdown-link .popavatar").click()
time.sleep(1)
driver.find_element(By.LINK_TEXT, "Logout").click()

```
